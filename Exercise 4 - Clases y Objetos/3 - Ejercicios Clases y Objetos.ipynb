## 1) Crear una clase básica con atributos y métodos básicos
Escribe un programa en Python para crear una clase Vehicle con atributos de instancia __max_speed__ y __mileage__. 

Y 2 métodos:
<ul>
    <li>get_max_speed()</li>
    <li>increase_mileage(ammount_to_increase)</li>
</ul>

```python
class Vehicle:
    def __init__(self, max_speed, mileage):
        self.max_speed = max_speed
        self.mileage = mileage

    def get_max_speed(self):
        return self.max_speed

    def increase_mileage(self, amount_to_increase):
        if amount_to_increase > 0:
            self.mileage += amount_to_increase
            print(f"Mileage increased by {amount_to_increase}.")
        else:
            print("Amount to increase must be greater than 0.")

car = Vehicle(120, 10000)
print("Car's max speed is {} and current mileage is {}".format(car.get_max_speed(), car.mileage))

car.increase_mileage(500)
print("The new car's mileage is {}".format(car.mileage))

print("Car's max speed is {} and current mileage is {}".format(car.get_max_speed(), car.mileage))
```

## 2) Crear una clase con atributos y métodos
Escribe una clase __Rectangle__ en lenguaje Python, que te permita construir un rectángulo con atributos de longitud y anchura.

Crea un método __perimeter()__ para calcular el perímetro del rectángulo y un método __area()__ para calcular el área del rectángulo.

Crear un método __display()__ que muestre la longitud, la anchura, el perímetro y el área de un objeto creado mediante una instanciación de la clase rectángulo.

Sobrescribir el método __\_str\_()__ para que la representación de la cadena por defecto de la clase sea la definida en display()

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def perimeter(self):
        return self.length * 2 + self.width * 2

    def area(self):
        return self.length * self.width

    def display(self):
        return f"Longitud: {self.length}, Anchura: {self.width}, Perimetro: {self.perimeter()}, Area: {self.area()}"

    def __str__(self):
        return self.display()

rec = Rectangle(5,15)
print(rec)
```

## 3) Crear una clase más avanzada con atributos y métodos
Crea una clase de Python llamada __BankAccount__ que represente una cuenta bancaria, teniendo como atributos 

<ul>
<li> <b>account_number</b> (de tipo numérico), <b>name</b> (nombre del titular de la cuenta, string), <b>balance</b>.</li>
</ul>

Y con los siguientes métodos:

<ul>
<li>Crear un constructor con parámetros: <b>account_number</b>, <b>name</b>, <b>balance</b>.</li>

<li>Crear un método <b>deposit(ammount)</b> que gestione las acciones de depósito.</li>

<li>Crear un método <b>withdrawal(ammount)</b> que gestione las acciones de retirada.</li>
    
<li>Crear un método <b>transfer(acc_to_transfer_to_obj, ammount)</b> que transfiera dinero a otro objeto de cuenta bancaria que reciba como parámetro.</li>

<li>Crear un método <b>apply_bank_fees(percentage)</b> que aplique las comisiones bancarias con un porcentaje del saldo actual de la cuenta.</li>
 
<li>Crear un método <b>display()</b> para mostrar los detalles de la cuenta.</li>
</ul>

Implementar también los métodos necesarios para comparar diferentes cuentas bancarias por saldo utilizando los operadores > < >= <= y ==.

```python
class BankAccount:
    """Clase que respresenta una Cuenta Bancaria basica"""
    def __init__(self, account_number, name, balance):
        """Constructor de la clase BankAccount"""
        self.account_number = account_number
        self.name = name
        self.balance = balance

    def deposit(self, ammount):
        """Deposita una cantidad positiva de dinero en la cuenta bancaria"""
        if ammount > 0:
            self.balance += ammount
        else:
            raise ValueError("Ammount must be >0")

    def withdrawal(self, ammount):
        """Retira una determinada cantidad positiva de dinero de la cuenta bancaria, si hay fondos suficientes"""
        if ammount > 0 and ammount <= self.balance:
            self.balance -= ammount
        else:
            raise ValueError("Ammount must be > 0 or Insufficient funds")
        
    def transfer(self, acc_to_transfer_to_obj, ammount):
        """Transfiere una cantidad positiva de dinero a otra cuenta bancaria, si hay fondos suficientes"""
        if ammount > 0 and ammount <= self.balance:
            self.balance -= ammount
            acc_to_transfer_to_obj.balance += ammount
        else:
            raise ValueError("Ammount must be > 0 or Insufficient funds")    

# Aqui tienes libertad para aplicar comisiones en cada operación, solo una vez por día, a varias cuentas, etc.
# Por simplificar el código haremos que la comisión se aplique solo una vez, independientemente del nº de operaciones que haga
    def apply_bank_fees(self, percentage = 0.5):
        """Aplica una tasa de un determinado porcentaje sobre el saldo actual de la cuenta"""
        fees = self.balance * (percentage/100)
        self.balance -= fees

    def display(self):
        """Muestra los detalles de la cuenta: numero de cuenta, nombre y balance"""
        return f"Account Number: {self.account_number}, Name: {self.name}, Balance: {self.balance}$"

    def __str__(self):
        """Devuelve el string indicado en display() cuando se instancia la clase"""
        return self.display()

    """Permite comparar diferentes cuentas bancarias segun su balance"""
    def __lt__(self, other_account):
        return self.balance < other_account.balance

    def __le__(self, other_account):
        return self.balance <= other_account.balance

    def __eq__(self, other_account):
        return self.balance == other_account.balance

    def __gt__(self, other_account):
        return self.balance > other_account.balance

    def __ge__(self, other_account):
        return self.balance >= other_account.balance

# Probando el codigo
main_account = BankAccount(147852, 'Dani', 1000)
second_account = BankAccount(798234, 'Sandra', 500)
print(main_account)

main_account.deposit(500)
print('Saldo tras el deposito:', main_account.balance)

main_account.withdrawal(100)
print('Saldo tras la retirada:', main_account.balance)

main_account.transfer(second_account, 300)
print('Saldo cuenta principal tras transferencia:', main_account.balance)
print('Saldo cuenta secundaria tras transferencia:', second_account.balance)

main_account.apply_bank_fees()
print('Saldo tras comisiones:', main_account.balance)

if main_account > second_account:
    print(f"Dani ({main_account.balance}$) tiene MAS dinero que Sandra ({second_account.balance}$)")
elif main_account < second_account:
    print(f"Dani ({main_account.balance}$) tiene MENOS dinero que Sandra ({second_account.balance}$)")
else:
    print(f"Dani ({main_account.balance}$) y Sandra ({second_account.balance}$) tienen el mismo dinero en el banco")
```

## 4) Herencia de clases

Crea una clase hija de Bus que herede de esta clase de Vehicle (diferente a la del primer ejercicio). La tarifa por defecto de cualquier vehículo es la capacidad de asientos * 100. Si el vehículo es una instancia de autobús, necesitamos añadir un 10% extra a la tarifa completa como cargo de mantenimiento. Así que la tarifa total para la instancia de autobús se convertirá en la cantidad final = tarifa total + 10% de la tarifa total.

Nota: La capacidad del autobús es de 50 asientos, por lo que el importe final de la tarifa debería ser de 5500. Necesita anular el método fare() de la clase Vehicle en la clase Bus.

Utiliza el siguiente código para su clase Vehicle padre.

```python
class Vehicle:
    def __init__(self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100

class Bus(Vehicle):
    def __init__(self, name, mileage, capacity):
        super().__init__(name, mileage, capacity)

    def fare(self):
        bus_fare = super().fare()     # Anular el metodo fare()
        return bus_fare * 1.10

school_bus = Bus("School Volvo", 12, 50)
print("Total Bus fare is:", school_bus.fare())
```

## 5) Ejercicio de excepciones

¡Vas a escribir una calculadora interactiva! Se supone que la entrada del usuario es una fórmula que consiste en un número, un operador (al menos + y -), y otro número, separados por espacios en blanco (por ejemplo, 1 + 1). Divide la entrada del usuario utilizando str.split(), y comprueba si la lista resultante es válida:

<ul>
<li>Si la entrada no consta de 3 elementos, lanza un FormulaError, que es una Excepción personalizada que tienes que crear y que sólo hereda de la clase de excepción base.</li>
<li>Intenta convertir la primera y tercera entrada en un float (así: float_value = float(str_value)). Captura cualquier ValueError que se produzca, y si esto pasa lanza un FormulaError</li>
<li>Si la segunda entrada no es '+' o '-', de nuevo lanza un FormulaError</li>
</ul>


Si la entrada es válida, realiza el cálculo e imprime el resultado. A continuación, se pide al usuario que proporcione una nueva entrada, y así sucesivamente, hasta que el usuario introduzca la salida.


Una interacción podría ser así:

```python
class FormulaError(Exception):
    """Define una nueva clase de excepción llamada 'FormulaError' que hereda de la clase base 'Exception'"""
    pass

def calculate(formula):
    """Define una funcion que toma una formula matematica como entrada y devuelve el resultado de la misma"""
    elements = formula.split()
    if len(elements) != 3:
        raise FormulaError("La formula tiene que tener extactamente 3 elementos separados por espacios")

    try:
        first_number = float(elements[0])
        operator = elements[1]
        second_number = float(elements[2])
    except ValueError:
        raise FormulaError("No se puede convertir el numero en float")

    if operator not in ["+", "-", "*", "/"]:
        raise FormulaError("El operador tiene que ser '+', '-', '*' o '/'")

    if operator == "+":
        return first_number + second_number
    elif operator == "-":
        return first_number - second_number
    elif operator == "*":
        return first_number * second_number
    else:
        if second_number == 0:
            raise FormulaError("No se puede dividir entre cero")
        else:
            return first_number / second_number

def main():
    """Define la funcion principal del programa. En un bucle infinito, solicita al usuario que ingrese una formula matematica"""
    while True:
        formula = input("Introduce una formula o calculo (si quieres salir, pulsa 's'): ")
        if formula.lower() == "s":
            print("Hasta la proxima!")
            break

        try:
            result = calculate(formula)
            print(f"El resultado de tu operacion es: {result}")
        except FormulaError as e:
            print(f"Error: {e}")

"""Si el script está siendo ejecutado directamente, se inicia el programa"""
if __name__ == "__main__":
    main()
```

## 6) Manejo de excepciones
Continúa con el ejercicio anterior y crea ahora tres clases hijas de FormulaError:

<ul>
<li><b>FormulaErrorNumber</b>: La levantarás cuando el usuario introduzca un valor erróneo para los números</li>
<li><b>FormulaErrorOp</b>: Se lanzará cuando el usuario introduzca un valor erróneo para el operador</li>
<li><b>FormulaErrorArgs</b>: Se lanzará cuando el usuario introduzca un número incorrecto de argumentos a la calculadora</li>
</ul>

Levántalos en los casos descritos y manéjalos con try, excepto para que puedas mostrar mensajes personalizados al usuario dependiendo de lo que esté mal en la entrada.

Puntos extra: Si también implementas la operación de división intenta manejar el caso ZeroDivisionError también

```python
class FormulaError(Exception):
    pass

class FormulaErrorNumber(FormulaError):
    """Error generado cuando se introducen valores no numéricos como operandos"""
    pass

class FormulaErrorOp(FormulaError):
    """Error generado cuando introduces un operador no aceptado"""
    pass

class FormulaErrorArgs(FormulaError):
    """Error generado cuando el número de argumentos es incorrecto (!= 3)"""
    pass

def calculate(formula):
    elements = formula.split()
    if len(elements) != 3:
        raise FormulaErrorArgs("La formula debe tener exactamente 3 elementos separados por espacios")

    try:
        first_number = float(elements[0])
        operator = elements[1]
        second_number = float(elements[2])
    except ValueError:
        raise FormulaErrorNumber("Ambos valores deben ser numéricos")

    if operator not in ["+", "-", "*", "/"]:
        raise FormulaErrorOp("El operador debe ser '+', '-', '*', o '/'")

    if operator == "+":
        return first_number + second_number
    elif operator == "-":
        return first_number - second_number
    elif operator == "*":
        return first_number * second_number
    else:
        if second_number == 0:
            raise ZeroDivisionError("No se puede dividir por cero.")
        else:
            return first_number / second_number

def main():
    while True:
        formula = input("Introduce una fórmula o cálculo (escribe 's' para salir): ")
        if formula.lower() == "s":
            print("Hasta la próxima!")
            break

        try:
            result = calculate(formula)
            print(f"El resultado es: {result}")
        except FormulaErrorNumber as e:
            print(f"Error en los numeros: {e}")
        except FormulaErrorOp as e:
            print(f"Error en el operador: {e}")
        except FormulaErrorArgs as e:
            print(f"Error en el número de argumentos: {e}")
        except ZeroDivisionError as e:
            print(f"Error al dividir por 0: {e}")

if __name__ == "__main__":
    main()
```
